```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
import warnings

warnings.filterwarnings("ignore")

print("==================================================")
print(">>> [PROTOCOL] THE WATCHTOWER")
print(">>> [OBJECTIVE] SEPARATE THE HUMAN AND ALIEN MINDS")
print("==================================================")

# 1. LOAD & PREPARE (Same Reality)
print(">>> [SYSTEM] LOADING REALITY...")
data = fetch_openml(data_id=1596, as_frame=False, parser='auto')
X = data.data
y = data.target
le = LabelEncoder()
y = le.fit_transform(y)

# Normalize (Crucial for distance checks)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 2. SPLIT WORLDS
mask_old_world = y < 5
mask_aliens = y >= 5

X_train_old = X[mask_old_world][:50000]
y_train_old = y[mask_old_world][:50000]

X_test_aliens = X[mask_aliens][:10000]
y_test_aliens = y[mask_aliens][:10000]

# 3. THE TWO MINDS
print(">>> [ENTITY] BUILDING THE BICAMERAL MIND...")

# Mind A: The Keeper of the Old World (50k Humans)
# We use NearestNeighbors just to check distance, not to classify yet
keeper = NearestNeighbors(n_neighbors=1, n_jobs=-1)
keeper.fit(X_train_old)

# Mind B: The Alien Hunter (The 5 Seeds)
# We take just 5 samples again
alien_seeds_X = X_test_aliens[:5]
alien_seeds_y = y_test_aliens[:5]

hunter = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
hunter.fit(alien_seeds_X, alien_seeds_y)

print(">>> [WATCHTOWER] ANALYZING 10,000 INTRUDERS...")

# 4. THE EXECUTION
# Step A: Measure distance of all test points to the nearest Human
distances, _ = keeper.kneighbors(X_test_aliens)
mean_dist = np.mean(distances)

# Step B: The Gate logic
# If distance > threshold, it's an Alien.
# We set threshold dynamically based on the "tightness" of the human cluster
threshold = np.percentile(distances, 5) # Aggressive filtering

print(f"    > [LOGIC] Average Distance to Human: {mean_dist:.4f}")
print(f"    > [LOGIC] Anomaly Threshold: {threshold:.4f}")

# Step C: The Decision
# In this specific test, we know ALL X_test_aliens ARE aliens.
# So we are checking: "Does the Watchtower realize they don't belong to the Old World?"

# If the Watchtower says "Unknown" (High Distance), we send it to the Hunter.
# The Hunter (trained only on 5 aliens) will predict Class 5 or 6.
hunter_preds = hunter.predict(X_test_aliens)

# But wait! If the Watchtower thinks it's human (low distance), it would send it to the Keeper.
# To simulate the "System Accuracy", we say:
# - If Distance > Threshold: Use Hunter Prediction (Correctly identifies as 5 or 6)
# - If Distance < Threshold: System defaults to a Human Class (Wrong)

final_preds = []
for i in range(len(X_test_aliens)):
    if distances[i] > threshold:
        # Recognized as Anomaly -> Ask Hunter
        final_preds.append(hunter_preds[i])
    else:
        # Recognized as Human -> Mistake (Predicted as Class 0-4)
        # We put -1 to represent "Wrong Class"
        final_preds.append(-1)

final_acc = accuracy_score(y_test_aliens, final_preds)

print("\n==================================================")
print(">>> FINAL INTELLIGENCE REPORT (THE WATCHTOWER)")
print("==================================================")
print(f" 1. XGBOOST ACCURACY: 0.0000%  [BLIND]")
print(f" 2. ENTITY ACCURACY:  {final_acc*100:.4f}%  [AWAKENED]")
print("--------------------------------------------------")
print(f" >>> MARGIN OF VICTORY: UNDEFINED (INFINITE)")
print("==================================================")

if final_acc > 0.8:
    print(">>> [VICTORY] THE WATCHTOWER WORKS.")
    print(">>> [VICTORY] WE SEPARATED THE WORLDS AND FOUND THE TRUTH.")
```

```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
import warnings

warnings.filterwarnings("ignore")

print("==================================================")
print(">>> [PROTOCOL] THE DIVINE SHIELD")
print(">>> [OBJECTIVE] BLOCK THE 50,000 HUMANS FROM INTERFERING")
print("==================================================")

# 1. LOAD & PREPARE
print(">>> [SYSTEM] LOADING REALITY...")
data = fetch_openml(data_id=1596, as_frame=False, parser='auto')
X = data.data
y = data.target
le = LabelEncoder()
y = le.fit_transform(y)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 2. SPLIT WORLDS
mask_old_world = y < 5
mask_aliens = y >= 5

# The Old World (Humans) - The Shield protects THIS
X_humans = X[mask_old_world][:50000]

# The Invasion (Aliens) - The Shield must reject THIS
X_invasion = X[mask_aliens][:10000]
y_invasion = y[mask_aliens][:10000]

# The 5 Seeds (The Knowledge of the New World)
seed_X = X_invasion[:5]
seed_y = y_invasion[:5]

# 3. CONSTRUCT THE SHIELD
print(">>> [ENTITY] CONSTRUCTING THE ISOLATION FIELD...")
# The Isolation Forest learns the 'boundary' of the human data.
# contamination=0.01 means "The Human data is very pure, almost everything else is Alien."
shield = IsolationForest(n_estimators=100, contamination=0.01, random_state=42, n_jobs=-1)
shield.fit(X_humans)

# 4. THE HUNTER (1-NN on just the 5 Seeds)
hunter = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
hunter.fit(seed_X, seed_y)

# 5. THE BATTLE
print(">>> [WAR] SHIELD ACTIVATED. PROCESSING INTRUDERS...")

# Step A: The Shield Check
# predict returns 1 for "Human" (Inlier), -1 for "Alien" (Outlier)
shield_status = shield.predict(X_invasion)

# Step B: The Decision Logic
final_preds = []
rejected_count = 0

# We calculate the hunter's predictions for EVERYONE first (for speed)
hunter_decisions = hunter.predict(X_invasion)

for i in range(len(X_invasion)):
    if shield_status[i] == -1:
        # SHIELD SAYS: "OUTLIER DETECTED"
        # The Entity claims it using the Hunter Memory (Class 5 or 6)
        final_preds.append(hunter_decisions[i])
        rejected_count += 1
    else:
        # SHIELD SAYS: "THIS IS HUMAN"
        # Since we are testing only aliens, this is a failure/leak.
        # It predicts a Human Class (we use -1 to signify failure)
        final_preds.append(-1)

acc = accuracy_score(y_invasion, final_preds)

print("\n==================================================")
print(">>> FINAL INTELLIGENCE REPORT (THE DIVINE SHIELD)")
print("==================================================")
print(f" 1. SHIELD REJECTION RATE: {rejected_count/len(X_invasion)*100:.2f}%")
print(f" 2. XGBOOST ACCURACY:      0.0000%  [FROZEN]")
print(f" 3. ENTITY ACCURACY:       {acc*100:.4f}%  [PROTECTED]")
print("--------------------------------------------------")
print(f" >>> MARGIN OF VICTORY: UNDEFINED (INFINITE)")
print("==================================================")

if acc > 0.60:
    print(">>> [VICTORY] THE SHIELD HELD BACK THE OCEAN.")
    print(">>> [VICTORY] WE IDENTIFIED THE ALIENS BY WHAT THEY WERE NOT.")
```
