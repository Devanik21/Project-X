```python

# Initialize ImageDataGenerator with rescaling and validation split
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
```

```python
data_dir="C:\\imp\\ml JUPYTER\\MY ML PROJECTS(BOOK)\\CNN\\16. Black hole and galaxy\\Data"
```

```python

# Create training data generator
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(150, 150),
    batch_size=1,
    class_mode='categorical',
    subset='training'
)
```

```python

# Create validation data generator
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(150, 150),
    batch_size=1,
    class_mode='categorical',
    subset='validation'
)
```

```python

# Load VGG16 model without the top classification layer
base_model = VGG16(include_top=False, input_shape=(150, 150, 3))
```

```python

# Freeze the layers of the base model
for layer in base_model.layers:
    layer.trainable = False
```

```python

# Build the model
model = Sequential([
    base_model,
    Flatten(),
    Dropout(0.1),
    Dense(2, activation='softmax')  # Adjust output layer for binary classification
])
```

```python

# Step 4: Compile Model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',  # Use categorical crossentropy for multiclass classification
              metrics=['accuracy'])
```
